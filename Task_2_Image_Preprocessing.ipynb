```python
from PIL import Image
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
```

```python
# Preprocessing pipeline
preprocess = transforms.Compose([
    transforms.Resize((224, 224)),        # Resizing to 224x224 pixels
    transforms.ToTensor(),                # Converting to tensor
    transforms.Normalize(                 # Normalizing for RGB
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    ),
])

```

```python
# Load and preprocess each image
image_paths = [
    "Sample1.png",
    "Sample2.png",
]

preprocessed_images = []

for path in image_paths:
    #Opening the image
    image = Image.open(path).convert("RGB")
    # Preprocess the image
    preprocessed_image = preprocess(image)
    preprocessed_images.append(preprocessed_image) #Inserting in empty list
    # Save the preprocessed image
    preprocessed_image_pil = transforms.ToPILImage()(preprocessed_image)
    preprocessed_image_pil.save(f"preprocessed_{path}")

# Display the preprocessed images for verification
def show_images(images):
    fig, axes = plt.subplots(1, len(images), figsize=(10, 5))
    for i, img in enumerate(images):
        img = img.permute(1, 2, 0)  # Convert from (C, H, W) to (H, W, C) for displaying
        axes[i].imshow(img)
        axes[i].axis("off")
    plt.show()

show_images(preprocessed_images)
```
